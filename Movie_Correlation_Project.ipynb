```python
#import Libraries
import pandas as pd
import seaborn as sns
import numpy as np


import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
#import data
df = pd.read_csv('/content/movies.csv')
```

```python
#Take a look at the data
df.head()
```

```python
#check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df = df.dropna()
```

```python
#check datatypes
df.dtypes
```

```python
#reformat budget column
df['budget']= df['budget'].astype('int64')
```

```python
df['gross']= df['gross'].astype('int64')
df.head()
```

```python
df['released'] = df["released"].astype(str)
```

```python
df
```

```python
df['correctYear'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
#order rows by gross column
df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
#look at all the data
#pd.set_option('display.max_rows', None)
```

```python
#Drop duplicates
df = df.drop_duplicates()
```

```python
#Extract month of release from the release column
df["Month"] = df["released"].str.split(" ").str[0]
```

```python
#Rename the Month column for clarity
df.rename(columns={'Month':'Release_Month'}, inplace=True)
```

```python
df["Release_Day"] = df["released"].str.split(" ").str[1]
```

```python
df['Release_Day'] = df['Release_Day'].str.replace(',',' ')
```

```python
#Drop the Released column since we already have the data splitted into other colunms
del df['released']
```

```python
df.head()
```

```python
#Check budget and company correlation
```

```python
#Scatter plot with budget Vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget Vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
#Budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color" : 'red'}, line_kws={'color': 'blue'})
```

```python
#Examine Correlation(only applicable to numeric data fields )
df.corr() #pearson
```

```python
df.corr(method = 'kendall') #kendal
```

```python
df.corr(method = 'spearman') #spearman
```

```python
df.corr(method='pearson') #pearson
```

```python
#Visualizing the correlation matrix from the pearson correlation
correlation_matrix = df.corr(method='pearson') #pearson
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
#Lets work on the company column
df_numerized = df

for col_name in df_numerized.columns:
  if(df_numerized[col_name].dtype == 'object'):
    df_numerized[col_name]= df_numerized[col_name].astype('category')
    df_numerized[col_name]= df_numerized[col_name].cat.codes

df_numerized
```

```python
#Visualizing the correlation for non-numeric data
correlation_matrix = df_numerized.corr(method='pearson') #pearson
sns.heatmap(correlation_matrix, annot=True)
plt.title('Numerized columns')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
#Check correlation pairs 
correlation_mat= df_numerized.corr()
Corr_pairs = correlation_mat.unstack()
Corr_pairs
```

```python
sorted_pairs = Corr_pairs.sort_values()
sorted_pairs
```

```python
high = sorted_pairs[(sorted_pairs)> 0.5]

high
```
